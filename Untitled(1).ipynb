```python
#no mistakes in data
df.isnull().sum()
```

```python
#rename weather column and move it to the second column

cols = list(df.columns.values)
df = df[[cols[0]] + [cols[-1]] + cols[1:7]]
df.rename(columns={'Weather':'Weather condition'})
df.head(2)
```

```python
#the mean of vsivility
y=df['Visibility_km'].mean()
print(y)
#standerd deviation
x=df['Visibility_km'].std()
print(x)
#variance og Rel Hum_%
v=df['Rel Hum_%'].var()
print(v)
```

```python
#9) Find all instances when 'Snow' was recorded.
Snow=df[df['Weather'].str.contains('Snow')]
Snow.head(10)

```

```python
#Find all instances when 'Wind Speed is above 24' and 'Visibility is 25'.
wind_s=df.loc[(df['Wind Speed_km/h']>24)&(df['Visibility_km']==25)]
wind_s.head(2)
```
